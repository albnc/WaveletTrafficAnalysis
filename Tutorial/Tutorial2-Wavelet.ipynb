```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
# z = (X - mu) / std
mean = 450
std = 150
n_hours = 24 * 30
rawSignal = np.random.randn(n_hours) * std + mean
x = np.arange(n_hours)
```

```python
plt.plot(x, rawSignal)
```

# Wavelet

1. Wavelet Family
2. Level of aggregation
3. Threshold

```python
maxL = 8
wp = pywt.WaveletPacket(data=rawSignal, wavelet='Haar', maxlevel=maxL)
```

```python
plt.plot(x, wp.data)
```

```python
def plotWavelet(wavePack, level):
    xi = np.arange(len(wavePack['a' * level].data))

    # Plot
    plt.subplot(2,1,1)
    plt.plot(xi, wavePack['a' * level].data)
    plt.title('Approximation ' + str(level))

    plt.subplot(2,1,2)
    plt.plot(xi, wavePack['d' * level].data)
    plt.title('Detail ' + str(level))

plotWavelet(wp, 5)
```

```python
rawSignal[np.random.choice(x, 15)] = 0
maxL = 8
wp = pywt.WaveletPacket(data=rawSignal, wavelet='Haar', maxlevel=maxL)

plotWavelet(wp, 5)
```
